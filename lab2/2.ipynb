```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import arviz as az
```

```python
np.random.seed(1)

x = []
for i in range(1, 10001):
    category = stats.multinomial.rvs(n=1, p=[0.25, 0.25, 0.3, 0.2], size=1)[0]
    latency = stats.expon.rvs(scale=1/4, size=1)
    if category[0] == 1:
        x.append((stats.gamma.rvs(a=4, scale=1/3, size=1) + latency)[0])
    elif category[1] == 1:
        x.append((stats.gamma.rvs(a=4, scale=1/2, size=1) + latency)[0])
    elif category[2] == 1:
        x.append((stats.gamma.rvs(a=5, scale=1/2, size=1) + latency)[0])
    else:
        x.append((stats.gamma.rvs(a=5, scale=1/3, size=1) + latency)[0])

print("P(X > 3) = ", (np.asarray(x) > 3).sum()/10000)
az.plot_posterior({'x': x})
plt.show()
```
