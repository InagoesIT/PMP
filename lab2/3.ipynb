```python
from scipy import stats
import matplotlib.pyplot as plt
import arviz as az
```

```python
# we can assume that every coin flipping is an independent event
# such that, the probability of getting 
# (head_head, head_tail, tail_head, tail_tail) will be:
# 0.5 * 0.3; 0.5 * 0.7; 0.5 * 0.3; 0.5 * 0.7, therefore
# the experiment can be modelled as a multinomial distribution

outcomes = stats.multinomial.rvs(n=10, p=[0.15, 0.35, 0.15, 0.35], size=100)
head_head = outcomes[:,0]
head_tail = outcomes[:,1]
tail_head = outcomes[:,2]
tail_tail = outcomes[:,3]

az.plot_posterior({'head_head':head_head, 'head_tail':head_tail, 'tail_head':tail_head, 'tail_tail':tail_tail})
plt.show()
```
