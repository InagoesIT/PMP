```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import arviz as az

from cmath import sqrt
```

```python
np.random.seed(1)

x = []
for i in range(1, 10001):
    if stats.binom.rvs(n=1, p=0.4, size=1) == 0:
        x.append(stats.expon.rvs(scale=1/4, size=1)[0])
    else:
        x.append(stats.expon.rvs(scale=1/6, size=1)[0])

az.plot_posterior({'x':x})
plt.show()
```

```python
mean = np.asarray(x).sum()/10000 # or = np.average(x)
print("mean = ", mean)
standard_deviation = sqrt(((np.asarray(x) - mean)**2).sum()/9999) #or = np.std(x)
print("standard deviation = ", standard_deviation)
```
